### Exploratory Data Analysis for all Data Roles
#### Roles to Explore

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title', fontweight='bold', pad=15)
plt.xlabel('Number of Jobs')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

### Countries to Explore

```python
df_plot = df['job_country'].value_counts().to_frame().head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,  x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country', fontweight='bold', pad=15)
plt.xlabel('Number of Jobs')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

#### Companies to Explore

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company', fontweight='bold', pad=15)
plt.xlabel('Number of Jobs')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

#### Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(12, 4))

for i, (column, title) in enumerate(dict_column.items()):
    counts = df[column].value_counts()
    ax[i].pie(
        counts, 
        labels=counts.index, 
        autopct='%1.1f%%', 
        startangle=90,
        colors=['#8ECAE6', '#219EBC']
    )
    ax[i].set_title(title, fontweight='bold')

plt.show()
```

### Exploratory Data Analysis for Data Analysts in the US

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

#### Filter for US Data Analyst roles

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
```

#### Locations to Explore

```python
df_plot = df_DA_US['job_location'].value_counts().to_frame().head(10)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()

plt.title('Counts of Job Locations for Data Analyst in the US', fontweight='bold', pad=15)
plt.xlabel('Number of Jobs')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

#### Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(12,4))

for i, (column, title) in enumerate(dict_column.items()):
    counts = df_DA_US[column].value_counts()
    ax[i].pie(
        counts, 
        labels=counts.index, 
        autopct='%1.1f%%', 
        startangle=90,
        colors=['#8ECAE6', '#219EBC']
    )
    ax[i].set_title(title, fontweight='bold')

plt.show()
```

#### Companies to Explore

```python
df_plot = df_DA_US['company_name'].value_counts().to_frame().head(10)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()

plt.title('Counts of Companies for Data Analyst in the US', fontweight='bold', pad=15)
plt.xlabel('Number of Jobs')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
